## Data Cleaning : Clean NaN/Null values

```python
df1 = df.copy()
df1.dropna(axis = 1,inplace = True)
```

```python
df1.columns
```

```python
df2 = df.copy()

null_percentage = df2.Product.isnull().sum() / len(df2)
if null_percentage > 0.70:
    df2.dropna(axis = 1,inplace = True)
df2.Product.isnull().sum()
```

```python
df2.loc[(df2.Price > 30) & (df2.Price <= 40),"Price"] = np.nan
```

```python
df2.isnull().sum()
```

```python
df2.info()
```

# Filling Mean, Median, &Mode

```python
df2 = df2.copy()
```

```python
mean_value = df2.Price.mean()

mean_value
```

```python
df2['Price'] = df2.Price.fillna(mean_value)
df2.isnull().sum()
```

```python
median_value = df2.Price.median()
median_value
```

```python
df2['Price'] = df2.Price.fillna(median_value)
df2.isnull().sum()
```

```python
mode_value = df2.Brand.mode()
print(type(mode_value))
```

```python
df2['Brand'] = df2.Brand.fillna('mode_value')
df2.isnull().sum()
```

# EDA-6 Identify outlier

```python
df6 = df.copy()
sns.boxplot(x = 'Brand',data = df6)
```

```python
df6 = df.copy()
sns.boxplot(x = 'Price', y = 'Brand',data=df6)
```

```python
Q1 = df.Price.quantile(.25)
Q3 = df.Price.quantile(.75)

print(Q1)
print(Q3) 
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_bound = Q1-1.5 * IQR 
upper_bound = Q3+1.5 * IQR
print(lower_bound)
print(upper_bound)
```

```python
outliers = df['Price'][(df.Price < lower_bound) | (df.Price > upper_bound)]
print(outliers)
```

```python
outliers.count()
```

# Data cleaning-4 : Trim the outliers

```python
df7 = df.copy()
df7 = df7[(df.Price >= lower_bound) & (df.Price <= upper_bound)]
```

```python
df7.Price.unique()
```

# Method-2 to find the outlier using z score
## z_score = (data-mean) / SD

```python
df8 = df.copy()
df8['z_score'] = (df['Price'] - df.Price.mean()) / df.Price.std()
```

```python
df8.head()
```

```python
outilers = df8['Price'][(df8.z_score < -3 ) | (df8.z_score > 3)]
print(outilers)
```

```python
df8 = df8[(df8.z_score >= -3) & (df8.z_score <= 3)]
df8.Price.unique()
```

# Clipping method for removing outliers

```python
df8['Price'] = df.Price.clip(lower = 20 , upper = 150)
df8.Price.unique()
```
